```python
from ckip_transformers.nlp import CkipWordSegmenter
from ckip_transformers.nlp import CkipPosTagger
from ckip_transformers.nlp import CkipNerChunker
```

```python
CKIP_WS=CkipWordSegmenter(model="bert-base",device=0)
CKIP_POS=CkipPosTagger(model="bert-base",device=0)
CKIP_NER=CkipNerChunker(model="bert-base",device=0)
```

##斷詞(WS)

```python
text_ws=CKIP_WS(text)
print(text_ws)
print(text)
```

##詞性(POS)

```python
#詞性標記
text_pos=CKIP_POS(text_ws)
```

```python
#篩詞性
#pos_filter=詞性
def filter_by_pos(sent,pos_filter):
  filter_pos_sent=[]
  for j,(word,pos) in enumerate(zip(sent[0],sent[1])):
    if pos[0] in pos_filter:
      filter_pos_sent.append(word)
  return filter_pos_sent
```

###N開頭

```python
#設定詞性為N
pos_filter_N=['N']

text_ws_pos=zip(text_ws,text_pos)
text_filter_N=[filter_by_pos(sent,pos_filter_N)for sent in text_ws_pos]
```

```python
#將詞性為N的詞整理到list_N裡面
list_N=[word for listlist in text_filter_N for word in listlist]
print(list_N)
```

###V開頭

```python
#設定詞性為V
pos_filter_V=['V']

text_ws_pos=zip(text_ws,text_pos)
text_filter_V=[filter_by_pos(sent,pos_filter_V)for sent in text_ws_pos]
```

```python
#將詞性為V的詞整理到list_V裡面
list_V=[word for listlist in text_filter_V for word in listlist]
print(list_V)
```

##命名實體(NE)


```python
text_ner=CKIP_NER(text)
```

```python
print(text_ner)
```

```python
for a in text_ner:
  print(a)
```

```python
#篩命名實體
#ner_filter=命名實體tag
def filter_by_ner(sent,ner_filter):
  filter_ner_sent=[]
  for j,word in enumerate(sent):
    if word.ner in ner_filter:
      filter_ner_sent.append(word.word)
  return filter_ner_sent
```

```python
filter_ner_sent=[]
for j,word in enumerate(text_ner):
  if word.ner in "PERSON":
    filter_ner_sent.append(word.word)
```

```python

def filter_by_ner(sent,ner_filter):
  filter_ner_sent=[]
  for j,word in enumerate(sent):
    if word.ner in ner_filter:
      filter_ner_sent.append(word.word)
  return filter_ner_sent
text_ner=CKIP_NER(text)
ner_filter_P=['PERSON']
text_ner_P=[filter_by_ner(sent,ner_filter_P)for sent in text_ner]
list_P=[]
for P_word in text_ner_P:
  if len(P_word)==0:
    continue
  for word in P_word:
    list_P.append(word)
```

###PERSON

```python
ner_filter_P=['PERSON']

text_ner_P=[filter_by_ner(sent,ner_filter_P)for sent in text_ner]
```

```python
#將所有tag=PERSON的詞整理到list_P裡面
list_P=[]
for P_word in text_ner_P:
  if len(P_word)==0:
    continue
  for word in P_word:
    list_P.append(word)
print(list_P)
```

###PERSON，GPE


```python
ner_filter_PG=['PERSON','GPE']

text_ner_PG=[filter_by_ner(sent,ner_filter_PG)for sent in text_ner]
```

```python
print(text_ner_PG)
```

```python
#將所有tag=PERSON和GPE的詞整理到list_PG裡面
list_PG=[]
for PG_word in text_ner_PG:
  if len(PG_word)==0:
    continue
  for word in PG_word:
    list_PG.append(word)
print(list_PG)
```

##詞性為N的文字雲
